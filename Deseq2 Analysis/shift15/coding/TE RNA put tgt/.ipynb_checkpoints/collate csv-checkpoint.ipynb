```python
import numpy as np
import pandas as pd
import csv
```

```python
filteredGeneList = pd.read_csv('../../../../Database/filteredGenesDetails_human_240118.txt',sep='\t')
acc2gene={}
for idx,row in filteredGeneList.iterrows():
    acc2gene[row['AccNum']] = row['GeneName']
```

```python
rna21 = pd.read_csv('../csv_raw/RNA_T2_against_T1.csv')
te21 = pd.read_csv('../csv_raw/TE_t2_t1.csv')
```

```python
te21.head()
```

```python
comb=pd.DataFrame({"AccNum": list(te21['Unnamed: 0'])})
```

```python
comb['Gene'] = comb['AccNum'].map(acc2gene)
comb['TElogFC'] = te21['log2FoldChange']
comb['RNAlogFC'] = rna21['log2FoldChange']
```

```python
comb.head()
```

```python
comb.to_csv('RNA TE T2T1.csv')
```
