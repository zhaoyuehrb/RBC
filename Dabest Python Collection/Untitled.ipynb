```python
import pandas as pd
import dabest
import glob
import matplotlib as mpl
import seaborn as sns
```

```python
infile = pd.read_csv("filteredGenesDetails_extendedAT.txt",sep=',')
infile.head()
```

```python
data = {'Others':infile['lastAT'][infile['Ery']==0], 'Erythroid':infile['lastAT'][infile['Ery']==1]}
df = pd.DataFrame(data) 
df.head()
```

```python
df_dabest = dabest.load(data = df,idx=("Others","Erythroid"),paired=False,resamples=20000)
plt=df_dabest.mean_diff.plot(swarmplot_kwargs={"size":0.6},swarm_label=("AT percentage"),dpi=300,es_marker_size=3);
plt.savefig('ATrich/'+'Erythroid_lastNT_dabest.png',dpi=300)
```

```python
data = {'Others':infile['allAT'][infile['Ery']==0], 'Erythroid':infile['allAT'][infile['Ery']==1]}
df = pd.DataFrame(data) 
df.head()
```

```python
df_dabest = dabest.load(data = df,idx=("Others","Erythroid"),paired=False,resamples=20000)
plt=df_dabest.mean_diff.plot(swarmplot_kwargs={"size":0.6},swarm_label=("AT percentage"),dpi=300,es_marker_size=3);
plt.savefig('ATrich/'+'Erythroid_allNT_dabest.png',dpi=300)
```
