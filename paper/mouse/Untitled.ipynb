```python
import tarfile
import glob
import gzip
import pandas as pd
import numpy as np
```

```python
tar = tarfile.open('GSE73357_RAW.tar')
tar.extractall(path='raw')
tar.close()
```

```python

tars = glob.glob('raw/*.gz')
for i,fname in enumerate(tars):
    #print(i)
    with gzip.open(fname, "rb") as f:
        data=f.read()
    with open(fname[:-3],'w') as f:
        f.write(data.decode('utf-8'))
```

```python
rpl10a = glob.glob('raw/*Rpl10a-Ribo*.txt')
ribos = glob.glob('raw/*Ribo*.txt')
ribos
```

```python
collate = pd.DataFrame()
collate['UCSC_ID'] = pd.read_csv(ribos[0],sep='\t')['UCSC_ID']
for idx,f in enumerate(ribos):
    df = pd.read_csv(f,sep='\t')
    group = f[15:-4]
    assert((df.columns == ['UCSC_ID','RPKM']).all())
    df.columns = ['UCSC_ID',group]
    collate = collate.merge(df, how='outer')
```

```python
collate
```

```python
collate_filter10 = collate.loc[(collate[collate.columns[1:]]>=10).all(1),:]
collate_filter10
```

```python
collate_filter10.to_csv('collate_rpkms.csv',index=False)
```

```python
for result in results:
    resdf = pd.read_csv(result)
    genes = list(resdf.loc[(resdf['log2FoldChange']<-0.64) & (resdf['padj']<0.05),'Unnamed: 0'])
    with open(result.split('\\')[1]+'_dep', 'w') as f:
        for item in genes:
            f.write("%s\n" % item)
```
